## Split

```python
txt = "School of Data Science"
x= re.split(" ",txt)
print(x)
```

```python
txt = "School of Data Science"
x= re.split("of",txt)
print(x)
```

```python
txt = "School-of-Data-Science"
x= re.split("-",txt)
print(x)
```

```python
txt = "School-of-Data-Science"
print(txt)
x= re.sub("-"," ",txt)
print(x)
```

```python
txt = "School-of-Data-Science"
print(txt)
x= re.sub("Data","Computer",txt)
print(x)
```

```python
# Remove leading zeros from an IP addresss
# input = 200.008.094.106
# output = 200.8.94.106

ip = "200.008.094.106"
pattern = re.compile(r'\b0+(?=\d)')
search_0 = re.sub(pattern,"",ip)
print(search_0)
```

```python
# Remove leading zeros from an IP addresss
# input = 200.008.094.106
# output = 200.8.94.106

ip = "200.008.094.106"
replace_0 = re.sub('[.]0*','.',ip)
print(replace_0)
```
