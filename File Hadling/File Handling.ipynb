# File Handling

## Read 

## Way 1:

```python
file = open('file1.txt','r')
for char in file:
    print(char)
```

## Way 2:

```python
file = open('file1.txt','r')
print(file.read())
```

## Way 3:

```python
file = open('file1.txt','r')
print(file.readline())
```

## Way 4:

```python
file = open('file1.txt','r')
print(file.read(2))
```

## Way 5:

```python
file = open('file1.txt','r')
print(file.readlines())
```

## Write

```python
file = open('file1.txt','w')
file.write("This is Statement writes one line into file..")
file.close()
```

```python
file = open('file1.txt','r')
print(file.readlines())
```

```python
file = open('file1.txt','a')
file.write("This line also added in this file")
file.close()
```

```python
file = open('file1.txt','r')
print(file.readlines())
```

```python
file = open('file1.txt','r')
print(file.read(6))

c= file.tell()
print(c)

file.seek(10)
print(file.read(5))

file.close()
```

```python
file = open('file1.txt','r')

count = 0
for char in file:
    count += len(char)
print("Number of Character is:->",count)
```
