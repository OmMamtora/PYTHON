```python
from tkinter import *
from tkinter import messagebox

def add():
    result = float(e1.get()) + float(e2.get())
    output_label.config(text=f"Result: {result}")

def subtract():
    result = float(e1.get()) - float(e2.get())
    output_label.config(text=f"Result: {result}")
    
def multiply():
    result = float(e1.get()) * float(e2.get())
    output_label.config(text=f"Result: {result}")
   
def divide():
    if float(e2.get()) == 0:
        messagebox.showerror("Error", "Cannot divide by zero")
    else:
        result = float(e1.get()) / float(e2.get())
        output_label.config(text=f"Result: {result}")
  

parent = Tk()
parent.title("Simple Calculator")
parent.geometry("300x300")

first_no = Label(parent, text="Enter first number:")
first_no.pack()

e1 = Entry(parent)
e1.pack()

sec_no = Label(parent, text="Enter second number:")
sec_no.pack()

e2 = Entry(parent)
e2.pack()

# Buttons
add = Button(parent, text="Add", command=add)
add.pack()

sub = Button(parent, text="Subtract", command=subtract)
sub.pack()

mul = Button(parent, text="Multiply", command=multiply)
mul.pack()

div = Button(parent, text="Divide", command=divide)
div.pack()

output_label = Label(parent, text="Result:")
output_label.pack()

parent.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox

def login():
    username = entry_user.get()
    password = entry_pass.get()

    if username == "" or password == "":
        messagebox.showwarning("Input Error", "Please enter both Username and Password")
    elif username == password:
        messagebox.showinfo("Login Successful", "Username and Password Match!")
    else:
        messagebox.showerror("Login Failed", "Username and Password do not match")


parent = Tk()
parent.title("Login Form")
parent.geometry("300x250")

# Username label and entry
user = Label(parent, text="Enter Username:")
user.pack()

entry_user = Entry(parent)
entry_user.pack()

# Password label and entry
password = Label(parent, text="Enter Password:")
password.pack()

entry_pass = Entry(parent, show="*")
entry_pass.pack()

# Login button
login = Button(parent, text="Login", command=login)
login.pack()

parent.mainloop()
```
