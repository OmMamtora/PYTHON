```python
### Leap year 

year = int(input("Enter Year:->"))

if year%4 == 0:
    if year%100 == 0:
        if year%400 == 0:
            print("Year is leap year..")
        else:
            print("Year is not leap year")
    else:
        print("leap year..")
else:
    print("Opps! Not leap year..")
```

```python
# Palindrom using loop
 
reverse = 0
palin = int(input("Enter Number:->"))
temp = palin

while palin > 0:
        digit = palin % 10
        reverse = (reverse * 10) + digit
        palin //= 10
        
print(reverse)
if temp == reverse:
    print("Palindrom")
else:
    print("Not Palindrom")
```

```python
# GCD

num1 = int(input("ENter Number 1:->"))
num2 = int(input("ENter number 2:->"))

while num2:
    num2 = num1%num2
    num1 = num2
    
    num1,num2 = num2, num1%num2
    
print(num1)
```
