```python
def array(n):
    add = 0 
    for x in n:
        add += x
    avg = add/len(n)
    return add,avg
```

```python
arr=[43,45,56,34,67,12,89,87,23]
a,b = array(arr)

print("Addition is ",a)
print(f'Average is {b:.2f}')
```

## Function with default value

```python
def mul(x=2,y=5,z=10):
    result = x*y*z
    return result
```

```python
print("Multi is",mul(12,3,6))
print("Multi is",mul(12,3))
print("Multi is",mul(12))
print("Multi is",mul())
print("Multi is",mul(z=2,y=3,x=7))
```
