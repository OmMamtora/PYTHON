## 1. Write a python program to create two NumPy arrays and perform element-wise addition, subtraction, multiplication, and division.

```python
arr1 = np.array([1,3,5,7,9])
arr2 = np.array([2,4,6,8,10])

print("Addition of arr1 and arr2 is ",np.add(arr1, arr2))
print("Subtraction of arr1 and arr2 is ",np.subtract(arr1, arr2))
print("Multiplication of arr1 and arr2 is ",np.multiply(arr1, arr2))
print("Division of arr1 and arr2 is ",np.divide(arr1, arr2))
```

# PANDAS

```python
import pandas as pd
```

## 1. Identify Missing Values Write a Python program using Pandas to: Load the dataset employee_performance.csv Display the total number of missing values in each column.

```python
df = pd.read_csv("employee_performance.csv")
df.head()

df.isnull().sum()
```

## 2. Replace Missing Numeric Values with Mean Replace missing values in numeric columns (Age, Salary, Rating, Attendance) with their mean values and display the updated dataset.

```python
df_mean = df[['Age', 'Salary', 'Rating', 'Attendance (%)']].fillna(df[['Age', 'Salary', 'Rating', 'Attendance (%)']].mean())
df_mean
```

```python
df_mean.isnull().sum()
```

## 3. Replace Missing Numeric Values with Median Replace missing values in numeric columns with their median and compare the result with the mean-based replacement.

```python
df_median = df[['Age', 'Salary', 'Rating', 'Attendance (%)']].fillna(df[['Age', 'Salary', 'Rating', 'Attendance (%)']].median())
df_median
```

```python
df_median.isnull().sum()
```

## 4. Replace Missing Categorical Values with Mode Replace missing values in categorical columns (Department, Gender) using their mode (most frequent value).

```python
df_mode = df[['Department','Gender']].fillna(df[['Department','Gender']]).mode().iloc[0]

df_mode
```

```python
df_mode.isnull().sum()
```

## 5. Fill Missing Values using a Fixed Value Replace all missing Rating values with a fixed value 3 (average rating).

```python
df_rating = df['Rating'].fillna(3)
df_rating
```

## 6. Forward Fill and Backward Fill Use forward fill (ffill) and backward fill (bfill) methods to handle missing data in the dataset. Compare the results.

```python
df_ffill = df.fillna(method = 'ffill')

df_bfill = df.fillna(method = 'bfill')
```

```python
df_ffill.isnull().sum()
```

```python
df_ffill.isnull().sum()
```
