# INSERT

```python
sql = "insert into Employee(name, id, salary, Dept_id) values (%s,%s,%s,%s)"

val = ("Om",122,35000.00,20)

cur.execute(sql,val)
myconn.commit()
```

```python
n = int(input("How many recoreds do you want to enter? "))
for i in range(n):
    name=input("Enter Employee Name :->")
    e_id = int(input("Enter Employee ID :->"))
    salary = int(input("Enter Salary :->"))
    d_id = int(input("Enter Department ID :->"))
    
    sql = "insert into Employee(name,id,salary,Dept_id) values (%s, %s, %s, %s)"
    val = (name,e_id,salary,d_id)
    
    cur.execute(sql,val)
    myconn.commit()
```
