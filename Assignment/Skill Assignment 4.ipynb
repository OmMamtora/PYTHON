## 2. Create a Python program to accept and add student education details in MySQL database table and display records of those students, who have secured first class in graduation. Assume suitable table structure.

```python
student_db = cur.execute("create table Students(roll_no int(5) not null primary key, name varchar(15) not null, course varchar(10) not null, year varchar(5) not null, percentage float(7) not null)")
```

```python
n = int(input("How many record do you want to add?"))

for i in range(n):
    eno = int(input("Enter Enrollment No :->"))
    name = input("Enter Name :->")
    course = input("Enter Course :->")
    year = input("Enter Year of study (1st,2nd,3rd) :->")
    percentage = float(input("Enter Percentage :->"))
    
    insert_sql = "insert into Students(roll_no,name,course,year, percentage) values (%s, %s, %s, %s, %s)"
    val = (eno,name,course,year,percentage)
    
    cur.execute(insert_sql,val)
    conn.commit()
```

```python
cur.execute("select * from Students where percentage > 80")

result = cur.fetchall()

for x in result:
    print(x)
```
