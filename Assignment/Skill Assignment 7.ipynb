## 1.Create a class Date with instance variables Day, Month, and Year. 
### Accept the values from user and initialize the instance variable through the constructors. 
### Overload the binary ‘-‘minus operator to find the difference between two Date class objects. 
### Create Python program to accept current date and birth date from user and find the current age of user.

```python
import datetime

class Date:
    def __init__(self,day,month,year):
        self.day = day
        self.month = month
        self.year = year
        
    def display(self):
        print("Year:",self.year)
        print("Months:",self.month)
        print("Days:",self.day)
        
    def __sub__(self,obj):
            
        if self.day < obj.day:
            self.month -=1
            self.day +=30
            
        if self.month < obj.month:
            self.year -=1
            self.month +=12
            
        
        years = self.year - obj.year
        months = self.month - obj.month
        days = self.day - obj.day
        
        return Date(days,months,years)
    
year1 = int(input("Enter Year of birth:->"))
month1 = int(input("Enter Month of birth:->"))
day1 = int(input("Enter Day of birth:->"))

year2 = int(input("Enter Current Year:->"))
month2 = int(input("Enter Current Month:->"))
day2 = int(input("Enter Current Day:->"))

date1 = Date(day1,month1,year1)
date2 = Date(day2,month2,year2)

date3 = date2 - date1
date3.display()
```

# 2. Write a class Volume with a function calculate():
## If one argument is passed → calculate volume of a cube.
## If two arguments are passed → calculate volume of a cuboid (length × breadth × fixed height = 1).
## If three arguments are passed → calculate volume of a cuboid (length × breadth × height).

```python
from multipledispatch import dispatch


class Volume:
    
    @dispatch(int)
    def display(self,cube):
        self.cube = cube
        return self.cube * self.cube * self.cube
    
    @dispatch(float,float)
    def display(self,lenght,breadth):
        self.length = lenght
        self.breadth = breadth
        self.height=1
        return self.length * self.breadth * self.height
    
    @dispatch(int,int,int)
    def display(self,length,breadth,height):
        self.length = length
        self.breadth = breadth
        self.height = height
        return length * breadth * height
        
v = Volume()
print(v.display(3))
print(v.display(3.5,5.0))
print(v.display(3,5,7))
```

# 3. Create a class MyString that stores a string.
## Overload the + operator to concatenate two strings.
## Overload the &gt; operator to check the string equality.
## Overload the == operator to check string equality.

```python
class MyString:
    def __init__(self,string):
        self.string = string
        
    def __add__(self,obj):
        return self.string + obj.string
    
    def __gt__(self,obj1):
        return len(self.string) > len(obj1.string)
    
    def __eq__(self,obj2):
        return self.string == obj2.string
    
    
my = MyString("Om")
my1 = MyString("Mamtora")

my3 = my+my1
print(my3)
print("Name 1 is greater then Name 2? :",my > my1)
print("Name 1 is equals to Name 2? :",my == my1)
```

## 4. Create an abstract base class called Shape with two instance variables Dimention- 1 and Dimention-2. Dimensions of shape used to compute the area. 
### Derive two specific classes called Triangle and Rectangle from the base Shape.
### Call the base class constructor through derived class constructor for initializing data members. 
### Add one member method area () to compute and display area of shapes. 
### Create python program to implement run time polymorphism by using method overriding.

```python
from abc import ABC,abstractmethod

class Shape:
    def __init__(self,Dimention1,Dimention2):
        self.dimen1 = Dimention1
        self.dimen2 = Dimention2
        
    @abstractmethod    
    def area(self):
        pass

class Triangle(Shape):
    def __init__(self,base,height):
        Shape.__init__(self,base,height)
        
    def area(self):
        return 0.5*self.dimen1* self.dimen2
    

class Rectangle(Shape):
    def __init__(self,length,breadth):
        Shape.__init__(self,length,breadth)
        
    def area(self):
        return self.dimen1 * self.dimen1
    
r = Rectangle(5,7)
t = Triangle(5,5)
print(r.area())
print(t.area())
```
