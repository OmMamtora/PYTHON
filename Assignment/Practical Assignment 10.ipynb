```python
from tkinter import *
from tkinter import messagebox
```

## 1. Create a python program that asks the user to typeÂ a word and return him its reverse as part of label.

```python
def reverse_string():
    input_text = e1.get()
    reversed_text = input_text[::-1]
    result_label.config(text=f"Reversed: {reversed_text}")

parent = Tk()
parent.title("Reverse String")
parent.geometry("300x300")

String = Label(parent, text="Enter a word:")
String.grid(row=0, column=0)

e1 = Entry(parent)
e1.grid(row=0, column=1)

submit = Button(parent, text="Validate", command=reverse_string)
submit.grid(row=1, column=0)

result_label = Label(parent, text="Reversed: ")
result_label.grid(row=2, column=0)

parent.mainloop()
```

## 2. Create a python program to display 3 checkbutton with name of 3 classes like FYBSC, SYBSC, and TYBSC. Display proper message when user select or unselect any of the checkbutton.

```python
def fun1():
    if checkBox1.get() == 1:
        messagebox.showinfo("Info","FYBSC Selected!")
    else:
        messagebox.showinfo("Info","FYBSC Unselected!")

def fun2():
    if checkBox2.get() == 1:
        messagebox.showinfo("Info","SYBSC Selected!")
    else:
        messagebox.showinfo("Info","SYBSC Unselected!")

def fun3():
    if checkBox3.get() == 1:
        messagebox.showinfo("Info","TYMSC Selected!")
    else:
        messagebox.showinfo("Info","TYBSC Unselected!")


parent = Tk()
parent.title("Check Button")
parent.geometry("300x300")

checkBox1 = IntVar()
checkBox2 = IntVar()
checkBox3 = IntVar()

checkbtn1 = Checkbutton(parent,text="FYBSC", variable = checkBox1, onvalue = 1, offvalue = 0, height = 2, width = 10, command = fun1)
checkbtn2 = Checkbutton(parent,text="SYBSC", variable = checkBox2, onvalue = 1, offvalue = 0, height = 2, width = 10, command = fun2)
checkbtn3 = Checkbutton(parent,text="TYBSC", variable = checkBox3, onvalue = 1, offvalue = 0, height = 2, width = 10, command = fun3)

checkbtn1.pack()
checkbtn2.pack()
checkbtn3.pack()

parent.mainloop()
```
