### 2. Create python program to read the content of one disk file, count the number of words from it and display back to user.

```python
count_words = 0
with open('file1.txt','r') as f1:
    for lines in f1:
        words = lines.split()
        count_words += len(words)
        
print("Total number of words:", count_words)
```

### 3. Create a python program to read the content of one file and copy into another file by replacing blank space by underscore (_). Display the content of both files at the end.

```python
f1=open("file1.txt",'r')
f2=open("file.txt",'w')

text=f1.read()
print(text)
for ch in text:
    if ch==' ':
        f2.write('_')
    else:
        f2.write(ch)
```

### 4. Create a python program that a defined function SWAP (x, y) to swap the values of two variables without using third variable. Accept the two numbers from user and pass it to functions.

```python
def swap(x,y):
    x = x + y
    y = x - y
    x = x - y

    return x,y

x=int(input("ENter number:->"))
y=int(input("Enter NUmber:->"))

swap(x,y)
```

### 5. Create a python program to create list of 20 numbers in it. Define a function FUN(num) which required list of 20 numbers as parameter. The function check which numbers are divisible by 3 and add it into separate list.

```python
def FUN(num):
    divisible_by_3 = []
    
    for n in num:
        if n % 3 == 0:
            divisible_by_3.append(n)
    print("Numbers divisible by 3:", divisible_by_3)


numbers = [i for i in range(1, 21)]

FUN(numbers)
```
