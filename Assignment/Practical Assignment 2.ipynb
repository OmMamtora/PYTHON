### 2. Create python program to read the content of one disk file, count the number of characters and digits from it and display back to user.

```python
with open("file1.txt") as f1:
    count = 0
    digit = 0
#     isdigit = [1,2,3,4,5,6,7,8,9,0]
                
    for line in f1:
        for char in line:
            length = len(char)
            count+=length
            
            if char.isdigit():
                    digit += 1

    print("Digit",digit)
    print("Count of character is",count)
```

### 3. Create python program to demonstrate the math module with any 5 functions of it.

```python
import math

# 1. Square root
num1 = 16
print("Square root of", num1, "is", math.sqrt(num1))

# 2. Power
base = 2
exp = 5
print(base, "raised to the power", exp, "is", math.pow(base, exp))

# 3. Factorial
num2 = 5
print("Factorial of", num2, "is", math.factorial(num2))

# 4. Floor
value1 = 7.8
print("Floor of", value1, "is", math.floor(value1))

# 5. Pi constant and using it to calculate the area of a circle
radius = 3
area = math.pi * math.pow(radius, 2)
print("Area of circle with radius", radius, "is", area)
```

### 4. Define a function to check whether given number is prime or not. Function should return Boolean value True or False.

```python
def prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Enter Number:-> "))

if prime(num):
    print(num,"is Prime number")
else:
    print(num,"is not Prime number")
```
