## 2. Write a Python program to accept a mobile number from user and check whether it is valid mobile number or not.

## it should be in 10 digit and it should starts with 6-9 followed by 9 digits, 
## country code +91

```python
def valid_mobile(x):
    pattern = r'^\+91[6-9]\d{9}$'
    if re.match(pattern, x):
        return 'Valid Mobile number'
    else:
        return 'Not Valid Mobile number please enter Valid mobile number..'
    
mob_no = input("Enter Mobile number:->")
print(valid_mobile(mob_no))
```

## 3. Write a Python program to accept an email id from user and check whether user has entered valid email ID or not.

### lower case letters are allowed 

### always starts with alphabets followed by alphanumeric characters _ and .

### @ symbol followed by alphabet and .

### numeric value allowed before @

```python
def email_verification(x):
    pattern = r'^[a-z][a-zA-Z0-9._]+@[a-z]*.[a-z.]+$'
    
    if re.match(pattern,x):
        return 'Valid Email id'
    else:
        return 'Not Valid Email id'

email = input("Enter Email_id:->")
print(email_verification(email))
```

## 4. Write a Python program to remove everything except alphanumeric characters from a string.
### Input text1 = &#39;**//Python Assignment// - 3. &#39;
### Output: PythonAssignment3

```python
def except_alphanumeric(x):
    pattern = r'[^a-zA-Z0-9]'
    clean_string = re.sub(pattern,'',x)
    if clean_string:
        return clean_string
    else:
        return None

name = '**//Python Assignment// - 3.'
print(except_alphanumeric(name))
```
