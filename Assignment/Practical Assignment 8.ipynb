## 1. Write a Python program to demonstrate the use of decorator to achieve function overloading in Python.

```python
from multipledispatch import dispatch

class Student:
        
    @dispatch()
    def display(self):
        print("No Details Provided..")
        print("----------------------")
        print()
    
    @dispatch(int)
    def display(self,prn):
        print("PRN of Student is ",prn)
        print("----------------------")
        print()
    
    @dispatch(int,str)
    def display(self,prn,name):
        print("PRN of Student is ",prn)
        print("Name of Student is ",name)
        print("----------------------")
        print()
        
    @dispatch(int,str,int)
    def display(self,prn,name,marks):
        print("PRN of Student is ",prn)
        print("Name of Student is ",name)
        print("Marks of Student is ",marks)
        print("----------------------")
        print()
        
stu = Student()
stu.display()
stu.display(20)
stu.display(20,"Om")
stu.display(20,"Om",98)
```

## 2. Write a Python program to demonstrate the operator overloading to achieve in operator overloading in Python. Overload any arithmetic operator.

```python
class OperatorOverloading:
    def __init__(self,a,b):
        self.num1 = a
        self.num2 = b
        
    def display(self):
        print("Number 1 :",self.num1)
        print("Number 2:",self.num2)
        print()
        
    def __mul__(self,obj):
        ob3 = OperatorOverloading(0,0)
        ob3.num1= self.num1*obj.num1
        ob3.num2 = self.num2*obj.num2
        return ob3
    
ob1 = OperatorOverloading(5,10)
ob2 = OperatorOverloading(4,8)

ob3 = ob1 * ob2

ob1.display()
ob2.display()
ob3.display()
```

## 3. Write a Python program to overload any relational operator to demonstrate the operator overloading in Python.

```python
class ChooseNumber:
    def __init__(self,n1):
        self.num1 = n1
        
    def display(self):
        print("Number is :",self.num1)
        
    def __eq__(self,obj):
        if self.num1 == obj.num1:
            return True
        else:
            return False
        
num1 = int(input("Enter Number 1:->"))
num2 = int(input("Enter Number 2:->"))

cn1 = ChooseNumber(num1)
cn2 = ChooseNumber(num2)

if cn1 == cn2:
    print("Entered Numbers are same..")
else:
    print("Entered number are not same..")
```

## 4. Write a Python program to demonstrate method overriding to achieve polymorphism in Python.

```python
from math import pi

class Shape:
    def __init__(self,name):
        self.name = name
        
    def area(self):
        pass
    
class Square(Shape):
    def __init__(self,lenght):
        super().__init__("Square")
        self.len = lenght
    
    def area(self):
        return self.len ** 2
    
class Circle(Shape):
    def __init__(self,radius):
        super().__init__("Cirlce")
        self.r = radius
        
    def area(self):
        return pi*self.r*self.r
    

C = Circle(5)
S = Square(5)
print(C.area())
print(S.area())
```

## 5. Write a Python program to demonstrate abstract class to achieve polymorphism in Python.

```python
from abc import ABC,abstractmethod

class Polygon(ABC):
    @abstractmethod
    def no_of_sides(self):
        pass
    
class Triangle(Polygon):
    def no_of_sides(self):
        print("I Have 3 Sides..")

class Pentagone(Polygon):
    def no_of_sides(self):
        print("I Have 5 Sides..")
        
class Hexagon(Polygon):
    def no_of_sides(self):
        print("I Have  Sides..")
        
class Quadrilateral(Polygon):
    def no_of_sides(self):
        print("I Have  Sides..")
        
        
t = Triangle()
t.no_of_sides()
print()

p = Pentagone()
p.no_of_sides()
print()

h = Hexagon()
h.no_of_sides()
print()

q = Quadrilateral()
q.no_of_sides()
```
