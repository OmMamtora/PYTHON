### Demonstrate the If-Else statement of Python by creating a python program to read a year from user, find whether that year is leap or not. (Conditions for being leap year: It should be divisible by four. If it’s a century year, it should be divisible by 400 also.)

```python
year = int(input("Enter Year:->"))

if year%4 == 0:
    if year%100 == 0:
        if year%400 == 0:
            print("Yes!!! Year is leap year..")
        else:
            print("Opps! Not leap year..")
    else:
        print("Yes!!! Year is leap year..")
else:
    print("Opps! Not leap year..")
```

### 2. Demonstrate the use of conditional statement by creating a python program to read a number from user, and check whether that number is palindrome or not. Display proper message back to user.


```python
reverse = 0
palin = int(input("Enter Number:->"))
temp = palin

while palin > 0:
        digit = palin % 10
        reverse = (reverse * 10) + digit
        palin //= 10
        
print(reverse)
if temp == reverse:
    print("Palindrom")
else:
    print("Not Palindrom")
```

### 5. Write a program to accept ‘n’ numbers from user store in list. Separate odd and even integers in separate lists.

```python
odd = []
even = []

n = int(input("How many numbers you want to add?"))

for i in range(n):
    num = int(input("Enter Number:->"))
    if num %2 !=0:
        odd.append(num)
    else:
        even.append(num)
        
print("Odd Numbers are:->",odd)
print("Even Numbers are:->",even)
```
