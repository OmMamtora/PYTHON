## 1. Create a class Product with Product ID, Name, Quantity, and Rate as instance variables. Accept the values of instance variables from user; calculate the Price of each product and Total Amount. Display all product details along with Prize per product and Total Amount back to user.

```python
class Product:
    
    total = 0
    
    def __init__(self,pId,name,quantity,rate):
        self.pID = pId
        self.name = name
        self.quantity = quantity
        self.rate = rate
        self.price = 0
        
    def cal(self):
        self.price = self.quantity * self.rate
        Product.total += self.price  
        
    def display(self):
        print("Product ID          :",self.pID)
        print("Product Name        :",self.name)
        print("Quantity of Product :",self.quantity)
        print("Rate Of Product     :",self.rate)
        print("Price Per Product   :",self.price)
        print("-------------------------------------")


n = int(input("How many data do you want to add?"))
for i in range(n):
    pid = int(input("Enter Product ID:"))
    pname = input("Enter Product Name:")
    quantity = int(input("Enter Quantity:"))
    rate = int(input("Enter Rate:"))
    print("--------------------------------")

    p = Product(pid,pname,quantity,rate)
    p.cal()
    p.display()
        

print("-------------------------------------")
print("Total Bill amount is ",Product.total)
```

## 4. Create a class BankAccount with Account Number, Customer Name, Account Type, and Balance as instance variables. 
### Provide methods to deposit () and withdraw (). Display account details after each transaction.

```python
class BankAccount:
    
    def __init__(self, accountNo,name,account_type,balance):
        self.accountNo = accountNo
        self.name = name
        self.account_type = account_type
        self.balance = balance
    
    def deposit(self):
        amount = int(input("Enter Amount to deposit: "))
        self.balance += amount
        print(f"{amount} deposited successfully!")
        self.display()
    
    def withdraw(self):
        amount = int(input("Enter Amount for withdraw:->"))
        
        if self.balance < 5000:
            print("Atleast maintain 5000RS. in account..")
            
        elif amount > self.balance:
            print("Don't have enough money in your account!")
            
        else:
            self.balance -= amount
            print(f"{amount} withdrawn successfully!")
            self.display()
    
    def display(self):
        print("Account No:",self.accountNo)
        print("Name      :",self.name)
        print("Type      :",self.account_type)
        print("Balance   :",self.balance,"\n")
        
        
accNo = input("Enter Account Number: ")
name = input("Enter Customer Name: ")
accType = input("Enter Account Type (Saving/Current): ")
balance = int(input("Enter Initial Balance (Minimum ₹5000): "))

bankAc = BankAccount(accNo, name, accType, balance)

while True:
    print()
    print("------ Menu ------")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Display Account Details")
    print("4. Exit\n")
    
    choice = input("Enter your choice: ")
    
    if choice == '1':
        bankAc.deposit()
    elif choice == '2':
        bankAc.withdraw()
    elif choice == '3':
        bankAc.display()
    elif choice == '4':
        print("Thank you..")
        break
    else:
        print("Invalid choice, please try again..")
        
```
