## 1. Create a class Vehicle with instance variable Vehicle Model, Registration number, Fuel capacity, and Vehicle Speed. It also contains following member functions in it:
### getVehicleDetails(): To accept vehicle details from user.
### showVehicleDetails(): To display vehicle details.
## Inherit two classes from Vehicle; class Truck with instance variable Weight limit and class Bus with instance variable Passenger capacity. Add appropriate member functions in both derived classes. Create Python program to solve this inheritance problem.

```python
class Vehicle:
    def getVehicleDetails(self):
        self.VehModel = input("Enter Vehicle Model No. :")
        self.rNo = int(input("Enter Vehicle Registration No : "))
        self.FCapacity = input("Enter Vehicle Fuel Capacity :")
        self.speed = int(input("Enter Vehicle Speed :"))
    
    def showVehicleDetails(self):
        print("---------------------------------------")
        print("Vehicle Model No. :",self.VehModel)
        print("Vehicle Registration No :",self.rNo)
        print("Vehicle Fuel Capacity :",self.FCapacity)
        print("Vehicle Speed :",self.speed)
    
class Truck(Vehicle):
    def __init__(self,weight_limt):
        self.weight_limit = weight_limt
        
        Vehicle.getVehicleDetails(self)
        Vehicle.showVehicleDetails(self)
        
    def tDisplay(self):
        print("Truck Weight Limit :",self.weight_limit)
        print()
        print("------------------------------------")
    
class Bus(Vehicle):
    def __init__(self,passenger_capacity):
        self.PCapacity = passenger_capacity
        
        Vehicle.getVehicleDetails(self)
        Vehicle.showVehicleDetails(self)
        
        
    def bDisplay(self):
        print("Bus Passenger Limit :",self.PCapacity)
        print()
        print("------------------------------------")

limit = int(input("Enter Passenger Bus Limit :"))
b = Bus(limit)
b.bDisplay()

weight = input("Enter Truck Weight Limit :")
t = Truck(weight)
t.tDisplay()
```

## 2. Create a class Employee with instance variable Employee No and Name.
## Also create a class Salary with instance variables Department and Basic Salary. 
## Derived a class EmployeeSalary from the above two classes. 
### Create Python program to calculate employee’s Gross Salary and Net Salary along with DA, HRA, PF, and IT. Accept all the information from user and Display all the salary details back to user.

```python
class Employee:
    def __init__(self,empNo,name):
        self.empNo = empNo
        self.name = name
    
class Salary:
    def __init__(self,dep,salary):
        self.dep = dep
        self.salary = salary
    
class EmployeeSalary(Employee, Salary):
    def __init__(self,empNo,name,dep,salary):
        
        Employee.__init__(self,empNo,name)
        Salary.__init__(self,dep,salary)
        
        
    def cal(self):
        self.DA = self.salary * 0.40
        self.HRA = self.salary * 0.20
        self.PF = self.salary * 0.12
        self.IT = self.salary * 0.10
        self.GrossSalary = self.salary + self.DA + self.HRA
        self.net_salary = self.GrossSalary - (self.PF+self.IT)
    
    def display(self):
        print()
        print("Employee No is :",self.empNo)
        print("Employee Name is :",self.name)
        print("Employee Department is :",self.dep)
        print("Employee  Basic Salary is :",self.salary)
        print()
        print("-------------------------------------------------")
        print()
        print("DA (40% of basic) of Employee is    :",self.DA)
        print("HRA (20% of basic) of Employee is   :",self.HRA)
        print("PF (12% of basic) of Employee is    :",self.PF)
        print("IT (10% of basic) of Employee is    :",self.IT)
        print("Total Gross Salary is               :",self.GrossSalary)
        print("Net Salary(After 10% tax) is        :",self.net_salary)

empno = int(input("Enter Employee No. :"))
name = input("Enter Name of Employee :")
dep = input("Enter Department of Employee :")
salary = int(input("Enter Salary of Employee :"))

empsal = EmployeeSalary(empno,name,dep,salary)
empsal.cal()
empsal.display()
```

## 3. Create a multilevel class structure: Student → Graduate → PostGraduate
### Student: roll_no, name
### Graduate: course_name, percentage
### PostGraduate: specialization, research_topic
### Display full academic details.

```python
class Student:
    def __init__(self,roll,name):
        self.prnNo = roll
        self.name = name
        
class Graduate(Student):
    def __init__(self,roll,name,cName,percentage):
        self.courseName = cName
        self.percentage = percentage
        
        Student.__init__(self,roll,name)
        
class PostGraduate(Graduate):
    def __init__(self,roll,name,cName,percentage,specialization,research_topic):
        self.specialization = specialization
        self.reTopic = research_topic
        
        Graduate.__init__(self,roll,name,cName,percentage)
        
    def display(self):
        print("PRN No. :",self.prnNo)
        print("Name :",self.name)
        print("Course Name :",self.courseName)
        print("Percentage :",self.percentage)
        print("Specialization :",self.specialization)
        print("Research Topic :",self.reTopic)
        
PG = PostGraduate(20,"Om","Msc.",95.50,"Data Science","ML")
PG.display()
```
