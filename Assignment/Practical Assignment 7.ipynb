## 1. Create a base class Student with attributes roll_no and name. Derive a class Marks to store marks of 3 subjects and a method to calculate the total and average.

```python
class Student:
    def __init__(self,rno,name):
        self.rollNo = rno
        self.name = name
        
class Marks(Student):
    def __init__(self,rno,name,m1,m2,m3):
        self.Python = m1
        self.Maths = m2
        self.DE = m3
        
        Student.__init__(self,rno,name)
        
    def cal(self):
        self.total = self.Python + self.Maths + self.DE
        self.average = self.total / 3
    
    def display(self):
        print("PRN No. :",self.rollNo)
        print("Name :",self.name)
        print("Marks of Python :",self.Python)
        print("Marks of Maths :",self.Maths)
        print("Marks of DE :",self.DE)
        
        print("Total is",self.total)
        print("Average is",self.average)
        
ma = Marks(20,"Om",98,97,99)
ma.cal()
ma.display()
```

## 2. Create a class Student with data members roll no &amp; name, add appropriate member methods to accept and display student details. 
### Inherit 3 classes from Student as Science, Arts &amp; Commerce. 
### Science class contains marks of two subjects as data members. 
### Arts class contains marks of two subjects as data members and 
### Commerce class contains marks of two subjects as data members. 
### Create Python program to achieve Hierarchical Inheritance.

```python
class Student:
    def __init__(self,rollno,name):
        self.prnNo = rollno
        self.name = name
        
class Science(Student):
    def __init__(self,rollno,name,sub1,sub2):
        self.Physics = sub1
        self.Biology = sub2
        
        Student.__init__(self,rollno,name)
        
    def cal(self):
        self.total = self.Physics + self.Biology
        
    def display(self):
        print("-------------------------------")
        print("Student From Science Stream..")
        print("PRN No. :",self.prnNo)
        print("Name :",self.name)
        print("Marks of Physics :",self.Physics)
        print("Marks of Biology",self.Biology)
        print("Total Of Marks :",self.total)
        
class Commerce(Student):
    def __init__(self,rollno,name,sub1,sub2):
        self.Account = sub1
        self.Statistics = sub2
        
        Student.__init__(self,rollno,name)
        
    def cal(self):
        self.total = self.Account + self.Statistics
        
    def display(self):
        print()
        print("-------------------------------")
        print("Student From Commerce Stream..")
        print("PRN No. :",self.prnNo)
        print("Name :",self.name)
        print("Marks of Account :",self.Account)
        print("Marks of Statistics :",self.Statistics)
        print("Total Of Marks :",self.total)
        
class Arts(Student):
    def __init__(self,rollno,name,sub1,sub2):
        self.Economics = sub1
        self.English = sub2
        
        Student.__init__(self,rollno,name)
        
    def cal(self):
        self.total = self.Economics + self.English
        
    def display(self):
        print()
        print("-------------------------------")
        print("Student From Arts Stream..")
        print("PRN No. :",self.prnNo)
        print("Name :",self.name)
        print("Marks of Economics :",self.Economics)
        print("Marks of English :",self.English)
        print("Total Of Marks :",self.total)

arts = Arts(1,"XYZ",76,78)
arts.cal()
arts.display()

comm = Commerce(2,"Om",98,99)
comm.cal()
comm.display()

sci = Science(3,"ABC",80,82)
sci.cal()
sci.display()
```

## Class Vehicle → attributes: vehicle_no, model, brand.
## Class Owner → attributes: owner_name, contact_no.
## Class VehicleOwnership inherits from both Vehicle and Owner.
### Add a method to display complete details of ownership.

```python
class Vehicle:
    def __init__(self,v_no,model,brand):
        self.vehicleNo = v_no
        self.model = model
        self.brand = brand

class Owner:
    def __init__(self,name,mob):
        self.oName = name
        self.phoneNo = mob
        
class VehicleOwnership(Vehicle,Owner):
    def __init__(self,v_no,model,brand,name,mob):
        Vehicle.__init__(self,v_no,model,brand)
        Owner.__init__(self,name,mob)
        
    def display(self):
        print("Vehicle No.",self.vehicleNo)
        print("Vehicle Model No.",self.model)
        print("Vehicle Brand ",self.brand)
        print("Name of Owner ",self.oName)
        print("Mobile No of Owner ",self.phoneNo)
        
VO = VehicleOwnership("GJ05","WDB 205","Mercedes-benz","Om",9874563210)
VO.display()
```

## Multilevel Inheritance: Person → Employee → Programmer
## Person: name, age
## Employee: emp_id, salary
## Programmer: programming_language, project_name
### Write methods to display all details and compute bonus for Programmer.

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
class Employee(Person):
    def __init__(self,name,age,empid,salary):
        self.empID = empid
        self.salary = salary
        
        Person.__init__(self,name,age)
        
    def eDisplay(self):
        print("Name of Employee :",self.name)
        print("Age of Employee :",self.age)
        print("Employee ID:",self.empID)
        print("Employee Salary:",self.salary)

class Programmer(Employee):
    def __init__(self,name,age,empID,salary,language,project_name):
        self.ProLanguage = language
        self.ProName = project_name
        
        Employee.__init__(self,name,age,empID,salary)
        
    def pDisplay(self):
        self.bonus = self.salary * 0.20
        print("Programming Language",self.ProLanguage)
        print("Project Name",self.ProName)
        print("Bonus for Programmer",self.bonus)

p = Programmer("Om",21,101,250000,"Python","Bank")
p.eDisplay()
p.pDisplay()
```
