### 1. Create a python program that calls 2 user defined function to find the GCD and LCM of given two numbers. Accept the two numbers from user and pass it to functions.


```python
def gcd(a,b):
    while b!= 0:
        a=b
        b=a%b
    return a

def lcm(a,b):
    g = gcd(a,b)
    lcm = (a*b)//g
    
    return lcm

num1 = int(input("Enter Number 1:->"))
num2 = int(input("Enter Number 2:->"))

g = gcd(num1,num2)
l = lcm(num1,num2)

print(f"GCD of {num1} and {num2} is :",g)
print(f"LCM of {num1} and {num2} is :",l)

```

### 2. Create a python program to read a decimal number from user and convert that number in binary and octal equivalent number.


```python
# Octal and Binary

# OCTAL
num = int(input("Enter Number:->"))
octal = ""
binary = ""

while num > 0:
    digit = num % 8
    octal = str(digit) + octal
    num //=8

print("Octal ",octal)


# BINARY
num = int(input("Enter Number:->"))
while num > 0:
    digit = num % 2
    binary = str(digit) + binary
    num //= 2
    
print("Binary ",binary)
```

### 3. Create a Python program to read n strings from user, store them into list. Display only those strings which is having even length


```python
n = int(input("How many strings do you want to add?"))

same_string = []

for i in range(n):
    string = input("Enter String:->")
    same_string.append(string)

for char in same_string:
    if len(char) % 2 == 0:
        print(char)
```

### 4. Accept data for 5 students with tuple including roll number, name, and marks. Find the topper from them.


```python
## Accept data for 5 students with tuple including roll number, name and marks Find topper from them

lst = []
max_marks = 0
rec = ()

for i in range(2):
    roll_No = int(input("Enter Roll No. :-> "))
    name = input("Enter Name :-> ")
    marks = int(input("Enter Marks :-> "))
    
    lst.append((roll_No,name,marks))
    
    if marks > max_marks:
        max_marks = marks
        rec=(roll_No, name, marks)

print("All Student records:->\n")
for i in lst:
    print(i)

print("PRN No. of Topper is :",rec[0]," Name is :",rec[1]," and Marks is :",rec[2])
```
