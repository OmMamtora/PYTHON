## 4. Write a Python program to match a string that contains only upper and lowercase letters, numbers, and underscores.

```python
def check(x):
    pattern = '[a-z_0-9A-Z]+'
    if re.search(pattern,x):
        return 'Found a match!'
    else:
        return 'Found a not match!'
    
print(check("This is Practical_Assignment 3 of PRN_NO. 250120020."))
print(check("PRN:250120020_"))
print(check("$#@$@%$@"))
print(check("OM Mamtora"))
```

## 5. Write a Python program that matches a string that has an a followed by one or more b&#39;s.

```python
def searching_string(string):
    
    pattern = 'ab+'
    
    if re.search(pattern,string):
        return 'Match!'
    else:
        return 'Not Match'

print(searching_string("ab"))
print(searching_string("abbbb"))
print(searching_string("a"))
print(searching_string("acb"))
```

## 6. Write a Python program to check for a number at the end of a string. 
### Input: abcdef Invalid
### abcdef6 Valid
### 23abc34def Invalid

```python
def end_number(string):
    pattern = '\d$'
    
    if re.search(pattern,string):
          return 'String ends with number'
    else:
        return 'String not ends with number'


print(end_number("Hellop2"))
print(end_number("Hellop"))
```

## 7. Write a Python program to find all five-character words in a string.
### Input = &#39;The quick brown fox jumps over the lazy dog.&#39;

```python
def find_words(x):
    pattern = r'\b\w{5}\b'
    return re.findall(pattern, x)

string = 'The quick brown fox jumps over the lazy dog.'
print(find_words(string))
```
