## 1. Create a class Employee with Emp ID, Name, Basic Salary as instance variables. Calculate DA (40% of basic), HRA (20% of basic), Gross Salary, and Net Salary (after 10% tax).
### Display employee salary slip.

```python
class Employee:
    
    def __init__(self,empId,name,basic_salary):
        self.empId = empId
        self.name = name
        self.basic_salary = basic_salary
    
    def cal(self):
        self.da = self.basic_salary * 0.40
        self.hra = self.basic_salary * 0.20
        self.gross_salary = self.basic_salary + self.da + self.hra
        self.net_salary = self.gross_salary * 0.90
    
    def display(self):
        print()
        print("------------------")
        print("Employee id is",self.empId)
        print("Employee Name is",self.name)
        print("Basic Salary of Employee is",self.basic_salary)
        print("DA (40% of basic) of Employee is",self.da)
        print("HRA (20% of basic) of Employee is",self.hra)
        print("Total Gross Salary is",self.gross_salary)
        print("Net Salary(After 10% tax) is",self.net_salary)

id = int(input("Enter Employee ID :->"))
name = input("Enter Employee Name :->")
basic_salary = int(input("Enter Basic salary Of Employee :->"))
emp = Employee(id,name,basic_salary)
emp.cal()
emp.display()
```

## 2. Create a class Book with Book ID, Title, Author, Price, and Quantity as instance variables.
### Accept values from user.
### Calculate total value of stock per book (Price × Quantity).
### Display all book details along with stock value.

```python
class Book:
    
    def __init__(self,bookid,title,author,price,quantity):
        self.bID = bookid
        self.title = title
        self.author = author
        self.price = price
        self.quantity = quantity
        
    def cal(self):
        self.stock = self.price * self.quantity

    def display(self):
        print("Book id is",self.bID)
        print("Title of book is",self.title)
        print("Author of book is",self.author)
        print("Price of book is",self.price)
        print("Quantity of book is",self.quantity)
        print("Stock Value",self.stock)
        print("--------------------------------")
        
n = int(input("How many records do you want to add?"))

for i in range(n):
    bid = int(input("Enter Book id :->"))
    title = input("Enter Book Title :->")
    author = input("Enter Book author :->")
    price = float(input("Enter Price of book :->"))
    quantity = int(input("Enter quantity of book :->"))
    print("--------------------------------")

    b = Book(bid,title,author,price,quantity)
    b.cal()
    b.display()
```

## 3. Create a class Room with Room Number, Customer Name, Days Stayed, Rate per Day.
### Calculate Total Bill = Days Stayed × Rate per Day.
### Display bill details for each customer.

```python
class Room:
    
    def __init__(self,rNO,name,stay,rate_per_day):
        self.rNo = rNO
        self.name = name
        self.stay = stay
        self.rate_per_day = rate_per_day
    
    def cal(self):
        self.total_bill = self.stay * self.rate_per_day
    
    def display(self):
        print()
        print("----------------------")
        print("Room NO.",self.rNo)
        print("Customer Name",self.name)
        print("Day Stayed",self.stay)
        print("Rate Per Day",self.rate_per_day)
        print("Total Bill",self.total_bill)

        
rno = int(input("Enter Room No. :->"))
name = input("Enter Customer Name :->")
stay_day = int(input("How Many days will the cusotmer stay :->"))
rate = int(input("Price of room per day :->"))
print()

r = Room(rno,name,stay_day,rate)
r.cal()
r.display()
```

## 4. Create a class CarRental with Car ID, Car Model, Rent per Day, and Number of Days. Calculate Total Rent = Rent per Day × Days.
### Display rental details for each customer.

```python
class CarRental:
    
    def __init__(self, cid, cModel, rent_per_Day, No_of_days):
        self.cID = cid
        self.cModel = cModel
        self.rent_per_day = rent_per_Day
        self.days = No_of_days
        
    def cal(self):
        self.total_rent = self.rent_per_day * self.days
    
    def display(self):
        print()
        print("----------------------------")
        print("Car ID :",self.cID)
        print("Car Model :",self.cModel)
        print("Rent of Car per day :",self.rent_per_day)
        print(f"Rent car for {self.days} days")
        print("Total Rent of car :",self.total_rent)
        
    
cid = int(input("Enter Car No. :->"))
cmodel = input("Enter Car Model :->")
rent = float(input("Enter Rent per Day :->"))
days = int(input("How many days you want to rent Car :->"))
print()


cr = CarRental(cid,cmodel,rent,days)
cr.cal()
cr.display()
```
