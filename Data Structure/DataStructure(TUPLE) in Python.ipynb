# TUPLE

```python
per = (67,85,92,76,99)
print(per)
type(per)
```

```python
per = per = (67,85,92.88,76,99)
print(per)
type(per)
```

```python
per = per = (67,85,"Om",76,99)
print(per)
type(per)
```

# Accessing Element

```python
per = per = (67,85,"Om",76,99)
print(per[2])
print(per[1:])
```

```python
## Accept data for 5 students with tuple including roll number, name and marks Find topper from them

lst = []

for i in range(3):
    roll_No = int(input("Enter Roll No. :-> "))
    name = input("Enter Name :-> ")
    marks = int(input("Enter Marks :-> "))
    
    lst.append((roll_No,name,marks))
    
std_data=tuple(lst)

print(std_data)

print(std_data[0][2])

maximum = std_data[0]

for j in maximum:
    xyz = max(std_data[j][2])
    print(xyz)
```
