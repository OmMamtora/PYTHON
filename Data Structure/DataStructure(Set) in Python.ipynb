## Operations Perform in Set

### UNION

```python
a={1,2,3,4,5}
b={3,4,5,6,7}
```

```python
print(a | b)
```

```python
a.union(b)
```

## INTERSECTION

```python
a={1,2,3,4,5}
b={3,4,5,6,7}
```

```python
print(a &  b)
```

```python
a.intersection(b)
```

## SET DIFFERENCE

```python
a={1,2,3,4,5}
b={3,4,5,6,7}
```

```python
print(a - b)
```

```python
print(b - a)
```

```python
a.difference(b)
```

```python
b.difference(a)
```
