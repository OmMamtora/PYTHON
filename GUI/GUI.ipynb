## Using CheckBox

```python
import tkinter.messagebox

def fun1():
    if checkBox1.get() == 1:
        messagebox.showinfo("Info","FYMSC Selected!")
    else:
        messagebox.showinfo("Info","FYMSC Unselected!")

def fun2():
    if checkBox2.get() == 1:
        messagebox.showinfo("Info","SYMSC Selected!")
    else:
        messagebox.showinfo("Info","SYMSC Unselected!")
```

```python
parent = Tk()
parent.title("Check Button")
parent.geometry("300x300")

checkBox1 = IntVar() # Define Empty Variable
checkBox2 = IntVar()

checkbtn1 = Checkbutton(parent,text="FYMSC", variable = checkBox1, onvalue = 1, offvalue = 0, height = 2, width = 10, command=fun1)
checkbtn2 = Checkbutton(parent,text="SYMSC", variable = checkBox2, onvalue = 1, offvalue = 0, height = 2, width = 10, command=fun2)

checkbtn1.pack()
checkbtn2.pack()

parent.mainloop()
```

## Using Radio Button

```python
def sel():
    if var.get()==1:
        str1 = "You selected the Option" + str(var.get())
    if var.get() == 2:
        str1 = "You selected the option" + str(var.get())

    label.config(text = str1)

root = Tk()
root.title("Radio Button")
root.geometry("300x300")

var = IntVar()

r1 = Radiobutton(root, text="Option 1", variable= var, value = 1, command= sel)
r1.pack()

r2 = Radiobutton(root, text="Option 2", variable = var, value=2, command=sel)
r2.pack()


label = Label(root)
label.pack()
root.mainloop()
```

```python
from functools import partial

def add(n1,n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(num1)+ int(num2)

    label_result.config(text = "Result :-> " + str(result))
    return 

root = Tk()
root.geometry("400x200")
root.title("Calculator")

number1 = StringVar()
number2 = StringVar()

labelNum1 = Label(root, text = "First Number")
labelNum1.grid(row = 1, column = 0)

labelNum2 = Label(root, text = "Second Number")
labelNum2.grid(row = 2, column = 0)

label_result = Label(root)
label_result.grid(row = 7, column = 2)

entryNum1 = Entry(root, textvariable = number1)
entryNum1.grid(row = 1, column = 2)

entryNum2 = Entry(root, textvariable = number2)
entryNum2.grid(row = 2, column = 2)

addition = partial(add, number1, number2)

buttonCal = Button(root, text="Addition", command = addition)
buttonCal.grid(row=3, column = 0)

root.mainloop()
```
