```python
class Person:
    def __init__(self,name,idnumber):
        self.name = name
        self.idnumber = idnumber
        
    def display(self):
        print(self.name)
        print(self.idnumber)
        
class Employee(Person):
    def __init__(self,idnumber,name,salary, post):
        self.salary = salary
        self.post = post
        
        Person.__init__(self,name,idnumber)
        
    def details(self):
        print("---------")
        print(self.name)
        print(self.idnumber)
        print(self.salary)
        print(self.post)
        
```

```python
emp = Employee(1,"Om",250000,"ML")
emp.display()
emp.details()
```

```python
class Base1(object):
    def __init__(self):
        self.str1="Om"
        print("Base1 Class")
    
class Base2(object):
    def __init__(self):
        self.str2 = "Mamtora"
        print("Base2 Class")
    
class Derived(Base1,Base2):
    def __init__(self):
        Base1.__init__(self)
        Base2.__init__(self)
        print("Derived Class")
    
    def printStr(self):
        print(self.str1, self.str2)
        
d = Derived()
d.printStr()
```

```python
class Student:
    def __init__(self,roll,name):
        self.rollNo = roll
        self.name = name
        
class Personal:
    def __init__(self,city,mob):
        self.city = city
        self.phoneNo = mob
        
class Marks(Student, Personal):
    def __init__(self,roll,name,city,mob,total):
        Student.__init__(self,roll,name)
        Personal.__init__(self,city,mob)
        
        self.total = total
        
    def show(self):
        print(self.rollNo)
        print(self.name)
        print(self.city)
        print(self.phoneNo)
        print(self.total)
```

```python
obj = Marks(1,"Om","Pune",9874563210,350)
obj.show()
```

```python
class Grandfather:
    def __init__(self,grandfather):
        self.grandfather = grandfather
        
class Father(Grandfather):
    
    def __init__(self,fathername,grandfather):
        
        self.fatherName = fathername
        Grandfather.__init__(self,grandfather)
        
class Son(Father):
    
    def __init__(self,sonname,fathername,grandfather):
        
        self.sonName = sonname    
        Father.__init__(self,fathername,grandfather)
        
    def print_Name(self):
        print("Grand",self.grandfather)
        print("Parent",self.fatherName)
        print("Son",self.sonName)

s = Son("Om","Vasudevbhai","Damarlal")
s.print_Name()
```

```python
class Publication:
    def __init__(self,pn,pr):
        self.pname = pn
        self.price = pr
        
class Book(Publication):
    def __init__(self,pn,pr,pg):
        Publication.__init__(self,pn,pr)
        self.pages = pg
        
    def bDisplay(self):
        print("Publication Name is",self.pname)
        print("Price is",self.price)
        print("Pages is",self.pages)

class Video(Publication):
    def __init__(self,pn,pr,pg,duration):
        self.duration_clip = duration
        Publication.__init__(self,pn,pr)
    
    def vDisplay(self):
        print("Duration is",self.duration_clip)
    
b = Book("BOB",450,120)
v = Video("NEW Video",699,300,"45 Minutes")
b.bDisplay()
v.vDisplay()
```
