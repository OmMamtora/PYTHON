```python
class Complex:
    def __init__(self,a,b):
        self.real = a
        self.img = b
        
    def show(self):
        print("Real :",self.real)
        print("Image :",self.img)
        
    def __add__(self,obj1):
        obj2 = Complex(2,3)
        obj2.real = self.real + obj2.real
        obj2.img = self.img + obj2.img
        return obj2

c1 = Complex(3.5,5.7)
c1.show()
c2 = Complex(3.5,5.7)
c1.show()

c2 = c1+c2
c2.show()
```

```python
class Number:
    def __init__(self,a):
        self.number = a
        
    def show(self):
        print("Value is",self.number)
        
    def __gt__(self,b):
        if self.number > b.number:
            return True
        else:
            return False
        
n1 = Number(10)
n2 = Number(20)

if n1 > n2:
    print("First Number is greater")
else:
    print("Second number is greater")
```

### Using Decorator

```python
from multipledispatch import dispatch

@dispatch(int,int)
def product(first,second):
    result = first * second
    print(result)
    
@dispatch(int,int,int)
def product(first,second,third):
    result = first * second * third
    print(result)
    
@dispatch(float,float,float)
def product(first,second,third):
    result = first * second * third
    print(result)

product(5,5)
product(5,5,5)
product(4.5,4.5,4.5)
```

```python
from multipledispatch import dispatch

class Student:
    def __init__(self, name="", age=0, marks=0):
        self.name = name
        self.age = age
        self.marks = marks

    # Overloading display with no arguments
    @dispatch()
    def display(self):
        print("No details provided")

    # Overloading display with 1 argument
    @dispatch(str)
    def display(self, name):
        print("Student Name:", name)

    # Overloading display with 2 arguments
    @dispatch(str, int)
    def display(self, name, age):
        print("Student Name:", name, "| Age:", age)

    # Overloading display with 3 arguments
    @dispatch(str, int, int)
    def display(self, name, age, marks):
        print("Student Name:", name, "| Age:", age, "| Marks:", marks)


# Example Usage
s = Student()

s.display()                       # No details provided
s.display("Alice")                 # Student Name: Alice
s.display("Bob", 20)               # Student Name: Bob | Age: 20
s.display("Charlie", 22, 85)       # Student Name: Charlie | Age: 22 | Marks: 85
```
