## Overridding

```python
class BASE:
    def __init__(self,num1):
        self.n1 = num1
        
    def show(self):
        print("Show Funciton of Base class..")
        print(self.n1)
        
class DERIVED(BASE):
    def __init__(self,num1,num2):
        BASE.__init__(self,num1)
        self.n2 = num2
        
    def show(self):
        print("Show Funciton of Derived class..")
        print(self.n2)
        super().show()
        
d1 = DERIVED(10,20)
d1.show()
```

```python
from math import pi

class Shape:
    def __init__(self,name):
        self.name = name
        
    def area(self):
        pass
    
class Square(Shape):
    def __init__(self,lenght):
        super().__init__("Square")
        self.len = lenght
    
    def area(self):
        return self.len ** 2
    
class Circle(Shape):
    def __init__(self,radius):
        super().__init__("Cirlce")
        self.r = radius
        
    def area(self):
        return pi*self.r*self.r
    

C = Circle(5)
S = Square(5)
print(C.area())
print(S.area())
```

```python
class CAT:
    def __init__(self,name):
        self.cat_name = name
    
    def display(self):
        print("Cat name is:",self.cat_name)
        print("I am a Cat")
        
class DOG:
    def __init__(self,name):
        self.dog_name = name
    
    def display(self):
        print("Dog name is:",self.dog_name)
        print("I am a Dog")
        
class RAT:
    def __init__(self,name):
        self.rat_name = name
    
    def display(self):
        print("Cat name is:",self.rat_name)
        print("I am a Rat")
        
        
def show(obj):
    print("Dynamic Binding..")
    obj.display()
    
    
    
c1 = CAT("ABC")
d1 = DOG("PQR")
r1 = RAT("XYZ")

show(c1)
print()
show(d1)
print()
show(r1)
        
```

```python
from abc import ABC,abstractmethod

class Polygon(ABC):
    @abstractmethod
    def no_of_sides(self):
        pass
    
class Triangle(Polygon):
    def no_of_sides(self):
        print("I Have 3 Sides..")

class Pentagone(Polygon):
    def no_of_sides(self):
        print("I Have 5 Sides..")
        
class Hexagon(Polygon):
    def no_of_sides(self):
        print("I Have  Sides..")
        
class Quadrilateral(Polygon):
    def no_of_sides(self):
        print("I Have  Sides..")
        
        
t = Triangle()
t.no_of_sides()
print()

p = Pentagone()
p.no_of_sides()
print()

h = Hexagon()
h.no_of_sides()
print()

q = Quadrilateral()
q.no_of_sides()
```
