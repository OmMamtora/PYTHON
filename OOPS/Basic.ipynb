```python
class Employee:
    def __init__(self,name,id):
        self.name = name
        self.id = id
        
    def display(self):
#         print("Id %d and Name is %s"%(self.id,self.name))
        print(f"ID: {self.id} and Name: {self.name}")
        
emp1 = Employee("Om",20)
emp2= Employee("XYZ",30)

emp1.display()
emp2.display()
```

```python
class Employee:
    comName = "Tata"
    
    def __init__(self,enam,eno):
        self.id = eno
        self.ename = enam
        
    def display(self):
        print("Company Name:",self.comName)
        print("Eno:",self.id)
        print("EName:",self.ename)
        
emp1=Employee("Om",20)
emp2=Employee("XYZ",35)

emp1.display()
emp2.display()
```

```python
class Student:
    
    def __init__(self,name,prn,m1,m2,m3):
        self.name = name
        self.prn = prn
        self.marks1 = m1
        self.marks2 = m2
        self.marks3 = m3
        
    def display(self):
        print("Name:",self.name)
        print("PRN:",self.prn)
        total = self.marks1+ self.marks2 + self.marks3
        per = total/3
        
        print("Total:",total)
        print("Precentage:",per)
        
s1 = Student("Om",20,23,21,22)
s1.display()        
```

```python
prn = int(input("Enter PRN:"))
name = input("Enter Name:")
marks1 = int(input("Enter Marks1:"))
marks2 = int(input("Enter Marks2:"))
marks3 = int(input("Enter Marks3:"))

s2 = Student(name,prn,marks1,marks2,marks3)
s2.display()
```
